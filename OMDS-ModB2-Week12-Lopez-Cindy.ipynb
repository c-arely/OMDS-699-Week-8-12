# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

The first approach, using two separate plots, makes it easier to compare patterns within each week individually. It reduces visual clutter and allows each trend to be viewed more clearly. However, it makes it harder to directly compare differences between the two weeks.
The second approach, plotting both weeks on the same graph, makes it easier to directly compare visitor patterns across weeks. This allows for quick identification of similarities and differences. However, it can become more cluttered and harder to read, especially if more time series are added.

```python
import os
import pandas as pd
import kagglehub
import matplotlib.pyplot as plt

# Download datasets
path_conversion = kagglehub.dataset_download("rabieelkharoua/predict-conversion-in-digital-marketing-dataset")

# Load dataset
conversion = pd.read_csv(os.path.join(path_conversion, "digital_marketing_campaign_dataset.csv"))

print(conversion.head())
```

```python

# Create bins for AdSpend
conversion["AdSpend_bin"] = pd.qcut(conversion["AdSpend"], 10)

# Average conversion by bin
trend = conversion.groupby("AdSpend_bin")["Conversion"].mean()

plt.figure(figsize=(8,5))
trend.plot()

plt.title("Conversion Rate Across Ad Spend Levels")
plt.xlabel("Ad Spend (Binned)")
plt.ylabel("Conversion Rate")

plt.xticks(rotation=45)
plt.show()
```

This visualization was designed to highlight the trend between advertising spend and conversion rate. By grouping ad spend into bins and plotting the average conversion rate, the chart simplifies a complex relationship into a clear trend. The line chart makes it easier to observe patterns across spending levels while reducing noise from individual data points, improving overall readability and storytelling.
